```python
import random
import plotly.graph_objs as go
import plotly.animation as animation

distance=50.
final_point1 = 0.
final_point2 = 100.
steps=0
distance_ls=[50]
steps_ls=[0]
frame=[]
while True:
  prob = random.random()
  if prob >= 0.5:
    distance += 1
    steps+=1
  else: 
    distance -= 1
    steps+=1
  distance_ls.append(distance)
  steps_ls.append(steps)
  frame.append(go.Frame(data = [go.Scatter(x=steps_ls,y=distance_ls)], layout=go.Layout(title="Drunken Man",xaxis_title="Steps",yaxis_title="Distance")))
  if distance == final_point1 or distance == final_point2:
    break
 
```

```python
fig = go.Figure(data = [go.Scatter(x=steps_ls,y=distance_ls)], layout=go.Layout(title="Drunken Man",xaxis_title="Steps",yaxis_title="Distance"), frames=frame)
fig.update_layout(updatemenus=[dict(type='buttons', showactive=False, buttons=[dict(label='Play', method='animate', args=[None, dict(frame=dict(duration=50, redraw=True), fromcurrent=True,transition=dict(duration=0))]),
dict(label='Pause', method='animate', args=[[None], dict(frame=dict(duration=0, redraw=True), mode='immediate', transition=dict(duration=0))])])])
fig.show()
```
